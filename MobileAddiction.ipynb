```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/mobile_addiction_factors_50cols.csv')
print(df.dtypes)
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Average_Screen_Time_per_Day'], kde=True, color='teal')
plt.title('Distribution of Average Screen Time per Day', fontsize=16)
plt.xlabel('Average Screen Time (Hours)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Anxiety_Level', y='Average_Screen_Time_per_Day', data=df, hue='Gender', palette='viridis')
plt.title('Average Screen Time vs Anxiety Level by Gender', fontsize=16)
plt.xlabel('Anxiety Level', fontsize=12)
plt.ylabel('Average Screen Time (Hours)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,8))
df['Favorite_App'].value_counts().plot.pie(autopct='%1.1f%%', startangle=140, cmap='Set3')
plt.title('Favorite App Distribution', fontsize=16)
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
correlation_matrix = df[['Average_Screen_Time_per_Day', 'Sleep_Hours_Per_Night', 'Academic_Performance_CGPA',
                         'Gaming_Hours_Per_Week', 'Screen_Time_on_Weekdays', 'Screen_Time_on_Weekends']].corr()

plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix for Key Features', fontsize=16)
plt.tight_layout()
plt.show()
```
